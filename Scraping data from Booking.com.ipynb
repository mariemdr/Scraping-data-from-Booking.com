# <font color='firebrick'> Scraping data from Booking.com </font>

### Dans ce projet, nous avons appliqué le web scraping sur le site Booking.com dans le but d'extraire des données nécéssaires à notre analyse.
###  Notre projet commence par l'extraction des vols de Tunis vers Paris en appliquant quelques filtres.Ensuite, nous sommes allés consulter et sauvegarder des données sur les hôtels 4 étoiles à Paris et avons finalement fait le passage obligatoire vers les destinations les plus célèbres  jusqu'à ce que nous ayons obtenu une réservation de vol complète.

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service 
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import pandas as pd
import time
from selenium.webdriver import ActionChains
from bs4 import BeautifulSoup
import requests
import csv
```

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.maximize_window()
driver.get('https://www.booking.com/')
time.sleep(1)
```

## <font color='darkorange'> Step 1:</font> <font color='royalblue'>Find a plane ticket to Paris </font>

```python
#Navigation to flight interface
flight=driver.find_element(By.XPATH,'//nav[@class="bui-tab bui-header__tab bui-tab--borderless bui-tab--light  bui-tab--rounded "]/ul/li[2]/a').click()
time.sleep(1)

#set informations

#1) From - To
from_tun=driver.find_element(By.XPATH,'//div[@class="search-form-inner"]/div/div[1]/div/div[1]/div/div/div').click()
time.sleep(1) 
from_input=driver.find_element(By.XPATH,'//input[@aria-label="Saisie de l’aéroport d’origine"]')
from_input.clear()
from_input.click()
time.sleep(1)
from_input.send_keys('Tunis')
time.sleep(1) 
pick_aireport=driver.find_element(By.XPATH,'//li[@aria-label="Aéroport Intl de Tunis-Carthage, Tunis, Tunisie, - Aéroport"]').click()
time.sleep(1)

destination=driver.find_element(By.XPATH,'//div[@class="search-form-inner"]/div/div[1]/div/div[3]/div/div/div').click()
time.sleep(1)
dest_input=driver.find_element(By.XPATH,'//input[@name="destination"]')
dest_input.send_keys('Paris')
dest_input.location_once_scrolled_into_view
time.sleep(2)
pick_airport=driver.find_element(By.XPATH,'//li[@data-apicode="PAR"]/div[2]/div').click()
time.sleep(1)

#2) dates
aller=driver.find_element(By.XPATH,'//div[@class="search-form-inner"]/div/div[1]/div/div[4]/div/div/div').click()
time.sleep(1)

date_depart=driver.find_element(By.XPATH,'//div[@aria-label="novembre 3"]').click()
time.sleep(3)
date_arrive=driver.find_element(By.XPATH,'//div[@aria-label="novembre 8"]').click()
time.sleep(2)

#3) person number
elem_scroll=driver.find_element(By.XPATH,'//div[@class="keel-grid "]')
elem_scroll.location_once_scrolled_into_view
adult=driver.find_element(By.XPATH,'//div[@class="keel-grid "]/div[2]/div/button').click()
time.sleep(1)
adult=driver.find_element(By.XPATH,'//div[@class="keel-grid v-c-p _ib1"]/div[3]/button').click()
time.sleep(1)

#4) search button
search_button=driver.find_element(By.XPATH,'//div[@class="search-form-inner"]/div/div[2]').click()
time.sleep(1)

#5) best choice
best_flight=driver.find_element(By.XPATH,'//a[@aria-label="Tri par Le meilleur choix - Ordre actuel : croissant - Appuyez pour trier par Le meilleur choix en ordre décroissant"]').click()
time.sleep(1)

#6) filtering by stops
elem_scroll1=driver.find_element(By.XPATH,'//div[@data-name="stops"]')
elem_scroll1.location_once_scrolled_into_view
time.sleep(2)
stops=driver.find_element(By.XPATH,'//div[@data-name="stops"]/div/div[3]/div[1]/li[1]')
action1=ActionChains(driver)
action1.move_to_element(stops).perform()
time.sleep(3)
button_only1=driver.find_element(By.XPATH,'//div[@data-name="stops"]/div/div[3]/div[1]/li[1]/div/div[1]/div/button').click()
time.sleep(2)

#7)filtering by airlines
elem_scroll2=driver.find_element(By.XPATH,'//div[@data-name="providers-section"]')
elem_scroll2.location_once_scrolled_into_view
time.sleep(3)
providers=driver.find_element(By.XPATH,'//div[@data-name="providers-section"]/div/div[3]/div[1]/li[6]')
time.sleep(3)
action1.move_to_element(providers).perform()
time.sleep(3)
button_only2=driver.find_element(By.XPATH,'//div[@data-name="providers-section"]/div/div[3]/div[1]/li[6]/div/div[1]/div/button').click()
time.sleep(3)


#8) get informations

#extraction de 105 vols
for i in range(6):
    but_scroll3=driver.find_element(By.XPATH,'//a[@class="moreButton"]')
    but_scroll3.location_once_scrolled_into_view
    time.sleep(3)
    but_scroll3.click()
    time.sleep(3)
    
time.sleep(3)

airline_depart=driver.find_elements(By.XPATH,'//div[@class="resultWrapper"]/div[1]/div/div[1]/div[2]/div/ol/li[1]/div/div/div[1]/div[2]')
Temps_depart_tunis=driver.find_elements(By.XPATH,'//div[@class="resultWrapper"]/div[1]/div/div[1]/div[2]/div/ol/li[1]/div/div/div[2]/div[1]')
Temps_arrivee_paris=driver.find_elements(By.XPATH,'//div[@class="resultWrapper"]/div[1]/div/div[1]/div[2]/div/ol/li[1]/div/div/div[4]/div[1]')
duree1=driver.find_elements(By.XPATH,'//div[@class="resultWrapper"]/div[1]/div/div[1]/div[2]/div/ol/li[1]/div/div/div[5]/div[1]')


airline_arrivee=driver.find_elements(By.XPATH,'//div[@class="resultWrapper"]/div[1]/div/div[1]/div[2]/div/ol/li[2]/div/div/div[1]/div[2]')
Temps_depart_paris=driver.find_elements(By.XPATH,'//div[@class="resultWrapper"]/div[1]/div/div[1]/div[2]/div/ol/li[2]/div/div/div[2]/div[1]')
Temps_arrivee_tunis=driver.find_elements(By.XPATH,'//div[@class="resultWrapper"]/div[1]/div/div[1]/div[2]/div/ol/li[2]/div/div/div[4]/div[1]')
duree2=driver.find_elements(By.XPATH,'//div[@class="resultWrapper"]/div[1]/div/div[1]/div[2]/div/ol/li[2]/div/div/div[5]/div[1]')
prix=driver.find_elements(By.XPATH,'//div[@class="resultWrapper"]/div[1]/div/div[2]/div/div[2]/div[1]/div/div[1]/div/a/div')

    
vols_list=[]
for i in range(len(airline_depart)):    
    vols_dict={'Airline_depart':airline_depart[i].text,
            'Temps_depart_tunis':Temps_depart_tunis[i].text,
            'Temps_arrivee_paris':Temps_arrivee_paris[i].text,
            'Duree_Tun_Paris':duree1[i].text,
            'Airline_arrivee':airline_arrivee[i].text,
            'Temps_depart_paris':Temps_depart_paris[i].text,
            'Temps_arrivee_tunis':Temps_arrivee_tunis[i].text,
            'Duree_Paris_Tun':duree2[i].text,
             'Prix':prix[i].text}
    vols_list.append(vols_dict)
    
    
time.sleep(2) 
```

```python
vols_data=pd.DataFrame(vols_list)    
vols_data   
```

## <font color='darkorange'> Step 3:</font> <font color='royalblue'>Search attractions </font>

```python
# Navigation to attractions interface
nav_scroll=driver.find_element(By.XPATH,'//nav[@class="bui-tab bui-header__tab bui-tab--borderless bui-tab--light  bui-tab--rounded "]')
nav_scroll.location_once_scrolled_into_view
time.sleep(3)
attraction=driver.find_element(By.XPATH,'//nav[@class="bui-tab bui-header__tab bui-tab--borderless bui-tab--light  bui-tab--rounded "]/ul/li[4]/a').click()
time.sleep(2)

# set informations

#1) destination
destination=driver.find_element(By.XPATH,'//input[@type="search"]')
destination.click()
time.sleep(2)
destination.send_keys('paris')
time.sleep(2)
pick_right_city_name=driver.find_element(By.XPATH,'//a[@class="css-17p3r2c"]').click()
time.sleep(3)

#2) calendar
calendrier=driver.find_element(By.XPATH,'//div[@class="css-17mdy6j"]').click()
time.sleep(1)
date_depart=driver.find_element(By.XPATH,'//span[@aria-label="3 novembre 2022"]').click()
time.sleep(1)
date_arrive=driver.find_element(By.XPATH,'//span[@aria-label="8 novembre 2022"]').click()
time.sleep(1)

#3) search button
button_rechercher=driver.find_element(By.XPATH,'//button[@class="fc63351294 a822bdf511 d4b6b7a9e7 cfb238afa1 f4605622ad css-14gytlh"]').click()
time.sleep(5)

#4) popular attractions
les_plus_populaires=driver.find_element(By.XPATH,'//label[@for="__bui-10-attr_book_score"]').click()
time.sleep(4)

#5) filtering by category
categorie=driver.find_element(By.XPATH,'//div[@class="css-crbgu0"]/div[1]/fieldset/div/div[1]/label').click()
time.sleep(4)

#6) filter by city
ville=driver.find_element(By.XPATH,'//div[@class="css-crbgu0"]/div[4]/fieldset/div/div[1]/label').click()
time.sleep(4)


#7) get informations about 100 destinations in Paris
dict_destinations={}
nb=0
i=1
while i<6:
    url=driver.current_url
    requests.get(url)
    soup=BeautifulSoup(driver.page_source,'html.parser')
    informations=soup.find_all('div',{'class':"ccff2b4c43 bdb988fcaf"})
    for info in informations:
        if info.find('div',{'class':"css-1qm1lh"}): 
            title=info.find('h4',{'class':"css-wqc2d7"}).text
            description=info.find('div',{'class':"css-1itiqpg"}).text
            price=info.find('div',{'class':"ac78a73c96"}).text
            nb+=1
            dict_destinations[nb]=[title,description,price]
        else:
            pass
    button=driver.find_element(By.XPATH,'//div[@role="group"]/div[3]/button').click()
    i=i+1
    time.sleep(1) 

```

```python
destinations_data=pd.DataFrame.from_dict(dict_destinations, orient = 'index', columns = ['Title','Description','Price'])
destinations_data 
```

```python
#8) save data in csv file
destinations_data.to_csv("Destinations_data.csv",index=False)
```
